```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define the layers
conv_layer = nn.Conv2d(in_channels=1, out_channels=3, kernel_size=3, stride=1, padding=1)
relu = nn.ReLU()
max_pool = nn.MaxPool2d(kernel_size=2, stride=2)
flatten = nn.Flatten()
dense_layer = nn.Linear(3 * 3 * 3, 64)
dense_activation = nn.ReLU()

# Example model wrapping the layers
class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.conv = conv_layer
        self.relu = relu
        self.pool = max_pool
        self.flatten = flatten
        self.fc = nn.Sequential(
            dense_layer,
            dense_activation
        )

    def forward(self, x):
        x = self.conv(x)
        x = self.relu(x)
        x = self.pool(x)
        x = self.flatten(x)
        x = self.fc(x)
        return x

# Instantiate the model
model = SimpleModel()

# Define the SGD optimizer
optimizer = optim.SGD(model.parameters(), lr=0.002, momentum=0.8)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchinfo import summary

# Define the layers
conv_layer = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, stride=1, padding=1)
relu = nn.ReLU()
max_pool = nn.MaxPool2d(kernel_size=2, stride=2)
flatten = nn.Flatten()
dense_layer1 = nn.Linear(32*100*100, 64)
dense_activation = nn.ReLU()
dense_layer2 = nn.Linear(64,1)

# Example model wrapping the layers
class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.conv = conv_layer
        self.relu = relu
        self.pool = max_pool
        self.flatten = flatten
        self.fc = nn.Sequential(
            dense_layer1,
            dense_activation,
            dense_layer2
        )
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.conv(x)
        x = self.relu(x)
        x = self.pool(x)
        x = self.flatten(x)
        x = self.fc(x)
        x = self.sigmoid(x)
        return x
# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Instantiate the model and move it to the appropriate device
model = SimpleModel().to(device)
# Display the model summary using torchinfo
model_summary = summary(model, input_size=(1, 3, 200, 200))
print(model_summary)
```

```python
import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from torchvision.datasets import ImageFolder
import os
```

```python
# Define the transformations for train and test datasets
train_transform = transforms.Compose([
    transforms.RandomRotation(50),               # Rotation
    transforms.RandomHorizontalFlip(),            # Horizontal flip
    transforms.RandomResizedCrop(200, scale=(0.9, 1.1)),  # Random zoom in/out
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),  # Shifting
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

test_transform = transforms.Compose([
    transforms.Resize((200, 200)),       # Resize image to 200x200
    transforms.ToTensor(),               # Convert image to tensor
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize image to [-1, 1]
])


# Load train and test datasets using ImageFolder (assuming data is organized in subdirectories for each class)
train_dataset = ImageFolder(
    root='./data/train/',       # Replace with the path to your training data
    transform=train_transform            # Apply the transformations defined above
)

test_dataset = ImageFolder(
    root='./data/test/',        # Replace with the path to your test data
    transform=test_transform             # Apply the transformations defined above
)

# Create DataLoader for training and testing
train_loader = DataLoader(
    train_dataset,
    batch_size=20,                      # Set batch size to 20
    shuffle=True                        # Shuffle the training data
)

test_loader = DataLoader(
    test_dataset,
    batch_size=20,                      # Set batch size to 20
    shuffle=True                        # Shuffle the test data
)


# To store training accuracy for each epoch
epoch_accuracies = []
epoch_losses = []
epoch_test_losses = []
running_loss=0.0
criterion = nn.BCELoss()
# Training loop
for epoch in range(10):  # Example: 10 epochs
    model.train()  # Set model to training mode
    correct = 0
    total = 0
    for images, labels in train_loader:
        # Move data to the same device as the model
        images = images.to(device)
        labels = labels.unsqueeze(1).float().to(device)  # Move labels to the same device as the model

        optimizer.zero_grad()  # Zero the gradients

        # Forward pass
        outputs = model(images)

        # Calculate loss
        loss = criterion(outputs, labels)  # Using BCELoss

        # Backward pass
        loss.backward()

        # Update model parameters
        optimizer.step()
        running_loss += loss.item()

        # Calculate accuracy
        predicted = (outputs > 0.5).float()  # Convert output to 0 or 1 based on sigmoid threshold
        correct += (predicted == labels).sum().item()  # Count correct predictions
        total += labels.size(0)  # Count total examples

    # Calculate training accuracy for the current epoch
    epoch_accuracy = correct / total * 100  # Accuracy as percentage
    epoch_accuracies.append(epoch_accuracy)
    # Average loss for the current epoch
    epoch_loss = running_loss / len(train_loader)
    epoch_losses.append(epoch_loss)

    print(f'Epoch [{epoch+1}/10], Loss: {epoch_loss:.4f}')

    print(f'Epoch [{epoch+1}/10], Loss: {loss.item():.4f}, Accuracy: {epoch_accuracy:.2f}%')
    model.eval()  # Set model to evaluation mode
    test_loss = 0.0
    with torch.no_grad():  # No gradient computation for validation
        for images, labels in test_loader:
            images = images.to(device)
            labels = labels.unsqueeze(1).float().to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)
            test_loss += loss.item()

    # Average test loss for the current epoch
    epoch_test_loss = test_loss / len(test_loader)
    epoch_test_losses.append(epoch_test_loss)
# After training is complete, calculate the median of the accuracies
median_accuracy = np.median(epoch_accuracies)
print(f'Median Training Accuracy: {median_accuracy:.2f}%')
std_loss = np.std(epoch_losses)
print(f'Standard Deviation of Training Loss: {std_loss:.4f}')
# Calculate the mean of the test losses over all epochs
mean_test_loss = np.mean(epoch_test_losses)
print(f'Mean Test Loss for All Epochs: {mean_test_loss:.4f}')
```
